```python
import random
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
n = 6
```

```python

random_numbers = [random.randint(1, n) for _ in range(100)]
print (random_numbers)

sns.countplot(x=random_numbers, palette="viridis")

# Add labels and title
plt.xlabel("Number")
plt.ylabel("Frequency")
plt.title("Frequency of Random Numbers (1-n)")

# Show the plot
plt.show()
```
