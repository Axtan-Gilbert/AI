What genre has the greatest number and percentage of tracks in top 100?

I want a sample of artists who hit the top 100 charts in 2024, not a name of every single artists ever so for that I will be doing a select group of people.

this dataset was put together by a single time but it is based off of old spotify streaming records, and data collected from him from other streaming platforms like apple music, youtube, tiktok, ect. Since this was put together by a single guy at the end of 2024 there could be errors in the data

```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

track_data_filename = 'Top Tracks Playlists - USA.csv'
track_data = pd.read_csv(track_data_filename)
df = track_data
```

United states graphs

```python
# Absolute frequency
freq_table = df["AI Genre"].value_counts().reset_index()
freq_table.columns = ["AI Genre", "Frequency"]

# Relative frequency (as proportion)
freq_table["Relative Frequency"] = freq_table["Frequency"] / freq_table["Frequency"].sum()

# Relative frequency as percentage
freq_table["Percentage"] = freq_table["Relative Frequency"] * 100

print(freq_table)
```

```python
# plotting bar chart
plt.figure(figsize=(10,6))
sns.barplot(x="Frequency", y="AI Genre", data=freq_table, palette="pastel")
plt.title("Frequency of Tracks by AI Genre (USA Top Tracks)")
plt.xlabel("Number of Tracks")
plt.ylabel("AI Genre")
plt.show()
```

```python
labels = freq_table["AI Genre"]
sizes = freq_table["Frequency"]

plt.figure(figsize=(8,8))
plt.pie(
    sizes, 
    labels=labels, 
    autopct="%1.1f%%",   # show percentages
    startangle=140,      # rotate start angle
    colors=sns.color_palette("pastel")  # pastel colors
)
plt.title("Relative Frequency of Tracks by AI Genre (USA Top Tracks)")
plt.show()
```

Global Chart Graphs

```python
#Defining Global Data -gdf
track_data2_filename = 'Top Tracks Playlists - Global.csv'
track2_data = pd.read_csv(track_data2_filename)
gdf = track2_data
```

```python
# Frequency
freq_table2 = gdf["AI Genre"].value_counts().reset_index()
freq_table2.columns = ["AI Genre", "Frequency"]

# Relative frequency (as proportion)
freq_table2["Relative Frequency"] = freq_table2["Frequency"] / freq_table2["Frequency"].sum()

# Relative frequency as percentage
freq_table2["Percentage"] = freq_table2["Relative Frequency"] * 100

print(freq_table2)
```

```python
# plotting bar chart
plt.figure(figsize=(10,6))
sns.barplot(x="Frequency", y="AI Genre", data=freq_table2, palette="husl")
plt.title("Frequency of Tracks by AI Genre (Global Top Tracks)")
plt.xlabel("Number of Tracks")
plt.ylabel("AI Genre")
plt.show()
```

```python
labels = freq_table2["AI Genre"]
sizes = freq_table2["Frequency"]

plt.figure(figsize=(8,8))
plt.pie(
    sizes, 
    labels=labels, 
    autopct="%1.1f%%",   # show percentages
    startangle=140,      # rotate start angle
    colors=sns.color_palette()
)
plt.title("Relative Frequency of Tracks by AI Genre (USA Top Tracks)")
plt.show()
```

For both charts the answer to the question is that Pop has the highest number of tracks in the top 100. In the United states pop had 32% of the top 100 and globally pop had 36% In Terms of actual numbers there were 18 pop songs on the global charts and 16 pop songs on the United states charts
